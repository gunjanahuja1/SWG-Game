```python
#import random
print('Winning rules of the game SNAKE WATER GUN are :\n'
      + "Snake vs Water -> Snake wins \n"
      + "Water vs Gun -> Water wins \n"
      + "Gun vs Snake -> Gun wins \n")

print('\n\nEach match wil be 5-points game')
```

```python
import random

def play_match():
    user = 0
    comp = 0

    for i in range(5):
        while True:
            print("Enter your choice:\n1 - Snake\n2 - Water\n3 - Gun")
            choice = int(input("Enter your choice: "))
            while choice < 1 or choice > 3:
                choice = int(input("Enter a valid choice (1-3): "))
            
            choices = ['Snake', 'Water', 'Gun']
            choice_name = choices[choice - 1]
            print("User choice is:", choice_name)

            comp_choice = random.randint(1, 3)
            comp_choice_name = choices[comp_choice - 1]
            print("Computer choice is:", comp_choice_name)

            if choice == comp_choice:
                print("It's a Draw. The match will be replayed.\n")
            else:
                break
        
        if (choice == 1 and comp_choice == 2) or (choice == 2 and comp_choice == 3) or (choice == 3 and comp_choice == 1):
            print("User wins!\n")
            user += 1
        else:
            print("Computer wins!\n")
            comp += 1
    
    print("Final Score:")
    print("User:", user)
    print("Computer:", comp)
    
    if user > comp:
        print('User wins')
    elif user < comp:
        print('Computer wins')
    else:
        print('It is a draw')
    
    if user > comp:
        return "user"
    elif user < comp:
        return "computer"
    else:
        return "draw"

def play_tournament():
    num_matches = int(input("Enter the number of matches in the tournament: "))
    user_tournament_score = 0
    comp_tournament_score = 0

    for match_num in range(1, num_matches + 1):
        print("\nMatch", match_num)
        print("---------------")
        print("Playing a match...")
        match_winner = play_match()
        if match_winner == "user":
            user_tournament_score += 1
            print("User wins the match!\n")
        elif match_winner == "computer":
            comp_tournament_score += 1
            print("Computer wins the match!\n")
        else:
            print("The match ended in a draw!\n")
        
        print("Tournament Score:")
        print("User:", user_tournament_score)
        print("Computer:", comp_tournament_score)

    if user_tournament_score > comp_tournament_score:
        print("\nYou win the tournament!")
    elif user_tournament_score < comp_tournament_score:
        print("\nComputer wins the tournament!")
    else:
        print("\nTournament ends in a Draw!")

# Main program
print("Winning rules of the game SNAKE WATER GUN are:")
print("Snake vs Water -> Snake wins")
print("Water vs Gun -> Water wins")
print("Gun vs Snake -> Gun wins")
print("\nEach match will be a 5-points game")

tournament_type = input("Do you want to play a tournament or a single match?\nEnter 'tournament' for multiple matches or 'single' for a single match: ").lower()

if tournament_type == "tournament":
    play_tournament()
elif tournament_type == "single":
    play_match()
else:
    print("Invalid choice. Please enter 'tournament' or 'single'.")
```
